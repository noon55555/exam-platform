```python
import pandas as pd
import nltk
nltk.download('punkt')
from nltk.tokenize import PunktSentenceTokenizer
```

```python
file = pd.read_csv('Tweets.csv')
```

```python
file.head()
```

```python
df = pd.DataFrame(file)
df = df[df['airline'] == 'United']

df = df[['airline_sentiment', 'text']]

df.head()
```

```python
df = df.reset_index(drop=True)

tokenizer = PunktSentenceTokenizer()

df['words_array'] = ''
df['sentences_array'] = ''

for row in range(df['text'].shape[0]):
  words = nltk.word_tokenize(df['text'][row])
  sentences = tokenizer.tokenize(df['text'][row])
  df['words_array'][row] = words
  df['sentences_array'][row] = sentences

df.head()
```

```python
df['words_count'] = df['words_array'].str.len()
df['sentences_count'] = df['sentences_array'].str.len()

df.head()
```
